````
# Outlier treatments
from scipy.stats import iqr

# Function to cap outliers based on the IQR
def cap_outliers(df, col):
    q1 = df[col].quantile(0.25)
    q3 = df[col].quantile(0.75)
    iqr_value = iqr(df[col])
    cutoff = iqr_value * 1.5
    lower, upper = q1 - cutoff, q3 + cutoff
    df[col] = np.clip(df[col], lower, upper)

# Apply to each numeric column
for col in numeric_cols:
    cap_outliers(hb_df, col)

# Replot to see the effect of outlier handling
plt.figure(figsize=(20, 20))
for i, col in enumerate(numeric_cols, 1):
    plt.subplot(num_rows, num_cols, i)
    sns.boxplot(data=hb_df, y=col, color='skyblue')
    plt.title(f'Boxplot - {col} After Outlier Capping')
    plt.ylabel(col)

plt.tight_layout()
plt.show()
````

````
# remove outliers
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from scipy.stats import iqr # Import the iqr function from scipy.stats

def count_and_cap_outliers(df, col):
    q1 = df[col].quantile(0.25)
    q3 = df[col].quantile(0.75)
    iqr_value = iqr(df[col])  # Now using the imported iqr function
    cutoff = iqr_value * 1.5
    lower_bound = q1 - cutoff
    upper_bound = q3 + cutoff

    # Count outliers
    outliers = df[(df[col] < lower_bound) | (df[col] > upper_bound)]
    num_outliers = outliers.shape[0]

    # Cap outliers
    df[col] = np.clip(df[col], lower_bound, upper_bound)

    return num_outliers
# Dictionary to store the number of outliers in each column
outliers_dict = {}

# Apply the function and store the count of outliers
for col in numeric_cols:
    outliers_count = count_and_cap_outliers(hb_df, col)
    outliers_dict[col] = outliers_count

# Print the number of outliers in each column
for col, count in outliers_dict.items():
    print(f'{col}: {count} outliers')

````

What all outlier treatment techniques have you used and why did you use those techniques?

Answer:

Interquartile Range (IQR) Method

*  Identifies and handles outliers based on the distance from the quartile range. Outliers are considered as data points lying below Q1 - 1.5 * IQR or above Q3 + 1.5 * IQR.
* It's robust and effective for skewed data, commonly used because it minimizes the influence of extreme values without assuming a specific data distribution

* After capping outliers, you replot boxplots for each column to visually assess the effect of your outlier treatment.

* capping outliers based on the Interquartile Range (IQR) method looks well-structured and is a good choice for reducing the effect of extreme values without removing data points entirely.

* This visualization uses a subplot grid defined by num_rows and num_cols, which should match or exceed the number of numeric columns to avoid errors.

Answer:

* Enhances Minority Class Representation:

SMOTE works by creating synthetic samples from the minority class (canceled bookings in this case) instead of creating exact copies. It does this by selecting examples that are close in the feature space, drawing a line between the examples in the feature space, and drawing a new sample at a point along that line. This approach helps to introduce more variety into the dataset, which is crucial for preventing overfitting that can occur with simple oversampling (repeating the same minority class instances).
* Balances the Dataset Without Losing Information:

Unlike undersampling, which can remove potentially valuable data from the majority class, oversampling with SMOTE increases the size of the minority class to match the majority class, preserving all original data and adding new, informative synthetic examples.

* Improves Model Performance:

By balancing the class distribution, models are less likely to be biased toward the majority class, improving performance metrics such as accuracy, precision, recall, and F1-score, especially for the minority class.

* Compatibility with Various Models:

SMOTE-generated synthetic samples are effective for both linear and non-linear classifiers, making this technique versatile across different types of predictive models.

### 7.Function For Confusion Matrix

````
import numpy as np
import matplotlib.pyplot as plt
from sklearn.metrics import confusion_matrix
import seaborn as sns

def plot_confusion_matrix(y_true, y_pred, classes, normalize=False, title=None, cmap=plt.cm.Blues):
    """
    This function prints and plots the confusion matrix.
    Normalization can be applied by setting `normalize=True`.
    """
    if not title:
        title = 'Confusion Matrix'
        if normalize:
            title = 'Normalized Confusion Matrix'

    # Compute confusion matrix
    cm = confusion_matrix(y_true, y_pred)

    # Normalize the confusion matrix if requested
    if normalize:
        cm = cm.astype('float') / cm.sum(axis=1)[:, np.newaxis]
        print("Normalized confusion matrix")
    else:
        print("Confusion matrix, without normalization")

    print(cm)

    fig, ax = plt.subplots()
    im = ax.imshow(cm, interpolation='nearest', cmap=cmap)
    ax.figure.colorbar(im, ax=ax)
    # We want to show all ticks...
    ax.set(xticks=np.arange(cm.shape[1]),
           yticks=np.arange(cm.shape[0]),
           # ... and label them with the respective list entries
           xticklabels=classes, yticklabels=classes,
           title=title,
           ylabel='True label',
           xlabel='Predicted label')

    # Rotate the tick labels and set their alignment.
    plt.setp(ax.get_xticklabels(), rotation=45, ha="right",
             rotation_mode="anchor")

    # Loop over data dimensions and create text annotations.
    fmt = '.2f' if normalize else 'd'
    thresh = cm.max() / 2.
    for i in range(cm.shape[0]):
        for j in range(cm.shape[1]):
            ax.text(j, i, format(cm[i, j], fmt),
                    ha="center", va="center",
                    color="white" if cm[i, j] > thresh else "black")
    fig.tight_layout()
    return ax
````

````
import numpy as np
import matplotlib.pyplot as plt
from sklearn.metrics import confusion_matrix
import seaborn as sns
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score, precision_score, recall_score, roc_auc_score, confusion_matrix, classification_report
from sklearn.impute import SimpleImputer  # Import SimpleImputer for handling NaNs


def logistic_regression_on_booking(data):
    # Define the feature set and target variable
    X = data.drop('is_canceled', axis=1)
    Y = data['is_canceled']

    # Select only numerical features for scaling
    numerical_features = X.select_dtypes(include=np.number).columns
    X_numerical = X[numerical_features]

    # Split the data into train and test set
    X_train, X_test, Y_train, Y_test = train_test_split(X_numerical, Y, test_size=0.2, random_state=1)
    print("The shape of training set is", X_train.shape, Y_train.shape)
    print("The shape of testing set is", X_test.shape, Y_test.shape)

    # Impute missing values using the mean strategy
    imputer = SimpleImputer(strategy='mean')  # Create an imputer instance
    X_train = imputer.fit_transform(X_train)  # Fit and transform on training data
    X_test = imputer.transform(X_test)  # Transform testing data using the trained imputer

    # Transforming the data
    scaler = StandardScaler()
    X_train = scaler.fit_transform(X_train)
    X_test = scaler.transform(X_test)

    # Build a logistic regression model
    log_reg = LogisticRegression(fit_intercept=True, max_iter=10000)
    log_reg.fit(X_train, Y_train)

    # Get the predicted classes and probabilities
    logistic_train_class = log_reg.predict(X_train)
    logistic_test_class = log_reg.predict(X_test)
    logistic_test_prob = log_reg.predict_proba(X_test)[:, 1]  # Probability for class '1' (canceled)

    # Calculating accuracy, precision, and recall on the test set
    logistic_test_acc = accuracy_score(Y_test, logistic_test_class)
    log_precision = precision_score(Y_test, logistic_test_class, average='weighted')
    log_recall = recall_score(Y_test, logistic_test_class, average='weighted')
    log_roc = roc_auc_score(Y_test, logistic_test_prob)

    # Output results
    print("\n")
    print("The accuracy on test dataset is", logistic_test_acc)
    print("The precision on test dataset is", log_precision)
    print("The recall on test dataset is", log_recall)
    print("The roc auc score on test dataset is", log_roc)
    print("\n")

    # Confusion matrix visualization
    cm = confusion_matrix(Y_test, logistic_test_class)
    sns.heatmap(cm, annot=True, fmt="d", cmap='Blues')
    plt.title('Confusion Matrix for Test Set')
    plt.ylabel('True Label')
    plt.xlabel('Predicted Label')
    plt.show()

    # Display classification report
    print("\n Classification report:")
    print(classification_report(Y_test, logistic_test_class))
    print("\n")

    # Prediction in a dataframe
    pred = pd.DataFrame({'Actual Label': Y_test,
                         'Predicted Label': logistic_test_class})
    print("The top 5 rows of actual vs predicted labels\n", pred.head())

    return logistic_test_acc, log_precision, log_recall, log_roc

````

````
import numpy as np
import pandas as pd
from sklearn.ensemble import RandomForestClassifier
from sklearn.model_selection import train_test_split, RandomizedSearchCV
from sklearn.metrics import accuracy_score, precision_score, recall_score, roc_auc_score, confusion_matrix, classification_report
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.preprocessing import LabelEncoder  # Import LabelEncoder
from sklearn.impute import SimpleImputer  # Import SimpleImputer

def random_forest_model(data):
    # Define features and target
    X = data.drop('is_canceled', axis=1)
    Y = data['is_canceled']

    # Convert categorical features to numerical using Label Encoding
    for col in X.select_dtypes(include=['object']).columns:
        le = LabelEncoder()
        X[col] = le.fit_transform(X[col].astype(str))

    # Impute missing values using SimpleImputer before splitting
    imputer = SimpleImputer(strategy='most_frequent')  # Use most_frequent for categorical features
    X = pd.DataFrame(imputer.fit_transform(X), columns=X.columns)


    # Split the data into train and test sets
    X_train, X_test, Y_train, Y_test = train_test_split(X, Y, test_size=0.2, random_state=1)
    print("The shape of training set is", X_train.shape, Y_train.shape)
    print("The shape of testing set is", X_test.shape, Y_test.shape)
    print("\n")


    # Define hyperparameters
    param_dict = {
        'n_estimators': [int(x) for x in np.linspace(100, 300, 3)],
        'max_depth': [int(x) for x in np.linspace(5, 10, 5)],
        'min_samples_split': [int(x) for x in np.linspace(5, 15, 3)],
        'min_samples_leaf': [int(x) for x in np.linspace(1, 5, 5)]
    }

    # Build the model and perform randomized search
    rf_model = RandomForestClassifier()
    rf_grid = RandomizedSearchCV(estimator=rf_model, param_distributions=param_dict, cv=5, verbose=2, scoring='accuracy')
    rf_grid.fit(X_train, Y_train)

    print(rf_grid.best_estimator_)
    rf_optimal_model = rf_grid.best_estimator_
    print(rf_grid.best_params_)

    # Predict and evaluate
    rf_train_class = rf_optimal_model.predict(X_train)
    rf_test_class = rf_optimal_model.predict(X_test)
    rf_test_prob = rf_optimal_model.predict_proba(X_test)[:, 1]

    rf_train_acc = accuracy_score(Y_train, rf_train_class)
    rf_test_acc = accuracy_score(Y_test, rf_test_class)
    rf_precision = precision_score(Y_test, rf_test_class, average='weighted')
    rf_recall = recall_score(Y_test, rf_test_class, average='weighted')
    rf_roc = roc_auc_score(Y_test, rf_test_prob)

    print("\nThe accuracy on train dataset is", rf_train_acc)
    print("The accuracy on test dataset is", rf_test_acc)
    print("The precision on test dataset is", rf_precision)
    print("The recall on test dataset is", rf_recall)
    print("The roc auc score on test dataset is", rf_roc, "\n")

    # Display confusion matrices
    print("Confusion matrix for train set:")
    sns.heatmap(confusion_matrix(Y_train, rf_train_class), annot=True, fmt="d", cmap='Blues')
    plt.show()

    print("Confusion matrix for test set:")
    sns.heatmap(confusion_matrix(Y_test, rf_test_class), annot=True, fmt="d", cmap='Blues')
    plt.show()

    return rf_test_acc, rf_precision, rf_recall, rf_roc
````

**Building A XG Boost Model**

````
import warnings
import pandas as pd
from sklearn.model_selection import train_test_split, RandomizedSearchCV
from sklearn.metrics import accuracy_score, precision_score, recall_score, roc_auc_score
from xgboost import XGBClassifier
import matplotlib.pyplot as plt

warnings.filterwarnings('ignore', category=UserWarning, message='.*use_label_encoder.*')

def xg_boost(X, Y):
    X_train, X_test, Y_train, Y_test = train_test_split(X, Y, test_size=0.2, random_state=42)

    parameters = {
        'n_estimators': [100, 150, 200],
        'max_depth': [3, 5, 7],
        'learning_rate': [0.01, 0.05, 0.1],
        'min_child_weight': [1, 2, 5]
    }

    model = XGBClassifier(eval_metric='logloss')
    xgb_search = RandomizedSearchCV(model, parameters, n_iter=10, scoring='accuracy', cv=3, verbose=0)
    xgb_search.fit(X_train, Y_train)

    best_model = xgb_search.best_estimator_
    predictions = best_model.predict(X_test)

    accuracy = accuracy_score(Y_test, predictions)
    precision = precision_score(Y_test, predictions, average='weighted')
    recall = recall_score(Y_test, predictions, average='weighted')
    roc_auc = roc_auc_score(Y_test, best_model.predict_proba(X_test)[:, 1])

    print(f"Accuracy: {accuracy}, Precision: {precision}, Recall: {recall}, ROC AUC: {roc_auc}")

    # Return the calculated metrics
    return accuracy, precision, recall, roc_auc

# Assuming X and Y are already defined from your dataset
# Now, the xg_boost function returns the metrics
xgb_metrics = xg_boost(X, Y)

# Accessing and storing the individual metrics
xgb_acc = xgb_metrics[0]
xgb_precision = xgb_metrics[1]
xgb_recall = xgb_metrics[2]
xgb_roc = xgb_metrics[3]
````
