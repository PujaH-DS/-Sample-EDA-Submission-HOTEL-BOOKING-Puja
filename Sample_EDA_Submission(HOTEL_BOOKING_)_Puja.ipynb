### What technique did you use to handle the imbalance dataset and why? (If needed to be balanced)

Answer:using the SMOTE (Synthetic Minority Over-sampling Technique).

1.Enhances Minority Class Representation:

SMOTE works by creating synthetic samples from the minority class (canceled bookings in this case) instead of creating exact copies. It does this by selecting examples that are close in the feature space, drawing a line between the examples in the feature space, and drawing a new sample at a point along that line. This approach helps to introduce more variety into the dataset, which is crucial for preventing overfitting that can occur with simple oversampling (repeating the same minority class instances).
2.Balances the Dataset Without Losing Information:

Unlike undersampling, which can remove potentially valuable data from the majority class, oversampling with SMOTE increases the size of the minority class to match the majority class, preserving all original data and adding new, informative synthetic examples.

3.Improves Model Performance:

By balancing the class distribution, models are less likely to be biased toward the majority class, improving performance metrics such as accuracy, precision, recall, and F1-score, especially for the minority class.

4.Compatibility with Various Models:

SMOTE-generated synthetic samples are effective for both linear and non-linear classifiers, making this technique versatile across different types of predictive models.
